```python
import plotly.graph_objects as go
import numpy as np

f = lambda x,y: 2*y**3 - 6*y**2 + 3*x**2*y
xrange = np.linspace(-5,5,50)
yrange = np.linspace(-5,5,50)

z1 = np.zeros((50,50))

for i in range(len(xrange)):
    for j in range(len(yrange)):
        z1[i][j] = f(xrange[i],yrange[j])


fig = go.Figure(data=[go.Surface(z=z1)])
fig.add_scatter3d(x = np.array([0,0]), y = np.array([0,2]), z = np.array([f(0,0),f(0,2)]), mode = 'markers')
fig.update_layout(scene = dict(
                    xaxis = dict(tickvals= [-5,5], nticks = 4),
                    yaxis = dict(tickvals= [-5,5], nticks = 4),
                    zaxis = dict(tickvals= [-5,5], nticks = 4))
                  )

fig.show()
```
